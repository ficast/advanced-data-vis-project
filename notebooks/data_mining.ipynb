```python
import geopandas as gpd

# Caminho do arquivo GeoJSON
geojson_path = "../project/geojson/br_states.json"

# Carregar o GeoJSON em um GeoDataFrame
gdf = gpd.read_file(geojson_path)

# Exibir as primeiras linhas
print(gdf.head())
```

```python
import matplotlib.pyplot as plt
print(gdf.columns)
gdf.plot()
plt.show()
```

```python

def plot_state(state_name):
    # Filtrar um estado (exemplo: "Minas Gerais")
    estado_escolhido = state_name
    gdf_estado = gdf[gdf["SIGLA"] == estado_escolhido]

    # Plotar apenas o estado escolhido
    fig, ax = plt.subplots(figsize=(6, 6))
    gdf_estado.plot(ax=ax, edgecolor="black", color="lightblue")
    plt.title(f"Mapa de {estado_escolhido}")
    plt.show()
```

```python
for state in gdf["SIGLA"]:
    plot_state(state)
```
